--------------------------------------------------------------------------------------------------------------------------

```python
data = data.drop(axis=0, index=0)
```

```python
data
```

-----------------------------------------------------------------------

```python
import openpyxl
from win32com import client
import os

def is_int(x):
    try:
        x = int(x)
        return 1
    except:
        return 0

i = 1
pdf_names = []
ref_names = []
while(is_int(data['Счет'][i])):
    print("ITER ", i)
    if(is_int(data['Задолжность'][i])):
        wb = openpyxl.load_workbook(filename = 'reference.xlsx')
        sheet = wb['Лист1']
        sheet['A2'] = adress
        sheet['B2'] = day
        sheet['C2'] = data['Счет'][i]
        sheet['D2'] = data['Квартира'][i]
        sheet['E2'] = data['Задолжность'][i]
        sheet['F2'] = deadline
        sheet['G2'] = street
        sheet['H2'] = house
        sheet['I2'] = corpus
        
        print("RESULT", sheet['D2'].value)
        
        wb.save(os.path.abspath(f'reference_{i}.xlsx'))
        wb.close()
        
        print("WRITE", i)
        
        excel = client.Dispatch("Excel.Application")
        sheets = excel.Workbooks.Open(os.path.abspath(f'reference_{i}.xlsx'))
        work_sheets = sheets.Worksheets[2]
        
        work_sheets.ExportAsFixedFormat(0, os.path.abspath(f'Квитанция_{i}.pdf'), OpenAfterPublish=False)
    
        sheets.Close(True)
        
        pdf_names.append(f'Квитанция_{i}.pdf')
        ref_names.append(f'reference_{i}.xlsx')
        print("CREATE FILE ", i)
    i += 1
```

```python
pdf_names
```

```python
from PyPDF2 import PdfFileMerger

merger = PdfFileMerger()

for pdf in pdf_names:
    merger.append(pdf)

merger.write("Квитанции.pdf")
merger.close()
```

```python
for pdf in pdf_names:
    try:
        os.remove(os.path.abspath(pdf))
    except:
        pass
```
